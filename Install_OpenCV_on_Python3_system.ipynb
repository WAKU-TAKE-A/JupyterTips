# Python3にOpenCVを入れる

[opencv-python](https://pypi.python.org/pypi/opencv-python)

を参考にしています。私はWinPython（本記事ではWinPython-64bit-3.5.3.1Qt5）を使用しました。

* 以下のセルを、必要があれば修正してください
    * WInPythonを「C:\WinPython3」インストールしている例です
    * OpenCV3.3.0.9をインストールする例です
* セルを実行してください（[Ctrl] + [Enter]）

DOSコマンドを、The jupyter notebookから実行できるなんて、すばらしい。  
本ページのipynbファイルは、説明書兼インストーラとして機能しています。

```python
%%cmd
cd /d C:\WinPython3\python-3.5.3.amd64
python -m pip install https://pypi.python.org/packages/5c/ff/11c585c89a74db70ff5c0916bbda5604564879fac9f08dbea456620aab98/opencv_python-3.3.0.9-cp35-cp35m-win_amd64.whl#md5=8e4214a4ccc6bd5bad4249e51c24691f
```

以下のデモスクリプトをjupyter上で実行するために、`lena-std.jpg`をアップロードします。

![upload01.jpg](upload01.jpg)

以下のセルを実行してください。

```python

%matplotlib inline
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("lena-std.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # OpenCVとmatplotlibのimshowでは、扱う画像配列の順番が異なる
plt.imshow(img)
plt.show()
```
